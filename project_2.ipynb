```python
# Библиотеки
import pandas as pd
```

```python
filename = r"https://github.com/Maksimka9/klustering/blob/main/full_data.xlsx?raw=true"
full_df = pd.read_excel(filename, index_col=False)
full_df
```

```python
pmi_df1 = full_df.loc[full_df['grouptext'].str.contains('ПМИ-.*', regex=True)]
pmi_df2 = pd.DataFrame(pmi_df1)
pmi_df2
```

```python
pmi_df2.groupby(['formobtext', 'subj']).aggregate({'mark':'mean'})
```
